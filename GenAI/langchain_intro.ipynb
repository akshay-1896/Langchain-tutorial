# Streaming & Batch

## Streaming

Streaming in Generative AI is the process of sending model-generated output incrementally (token-by-token) to the user in real time instead of delivering the entire response at once.

```python
model.invoke("Write me a 500 words paragraph on AI")
```

```python
for chunk in model.stream("Write me a 1000 words essay on AI"):
    print(chunk.text, end="", flush=True)
```

## Batching

Batching is the process of grouping multiple inputs and processing them together in a single request to improve computational efficiency and throughput.

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplanes fly?",
        "What is quantum computing?",
        "What is Deep Learning?"
    ]
)

for response in responses:
    print(response.content)
```

```python
# Restriction for number of queries at a time

responses = model.batch(
    ["Why do parrots have colorful feathers?",
     "How do airplanes fly?",
     "What is quantum computing?",
     "What is Deep Learning?"],
     config={
        'max_concurrency': 2, #Limit to 2 parallel calls
     }
)

for response in responses:
    print(response.content)
```
